```python
# Create a Pandas DataFrame with some example data
data = {'x': [1, 2, 3, 4, 5],
        'y1': [2, 4, 6, 8, 10],
        'y2': [100, 200, 300, 400, 500]}
df = pd.DataFrame(data)

# Create a 4 by 4 array of subplots
fig, axs = plt.subplots(4, 4, figsize=(12, 12))

# Loop through the subplots and plot data from the DataFrame in each one
for i in range(4):
    for j in range(4):
        ax = axs[i,j]
        ax.plot(df['x'], df['y1'], color='blue')
        ax.set_xlabel('X Label')
        ax.set_ylabel('Y1 Label', color='blue')
        ax.tick_params(axis='y', labelcolor='blue')
        
        ax2 = ax.twinx()
        ax2.plot(df['x'], df['y2'], color='red')
        ax2.set_ylabel('Y2 Label', color='red')
        ax2.tick_params(axis='y', labelcolor='red')
        
        ax.set_title('Subplot %d,%d' % (i,j))

# Adjust the layout and padding
fig.tight_layout(pad=1.0, w_pad=1.0, h_pad=1.0)

# Show the plot
plt.show()
```
