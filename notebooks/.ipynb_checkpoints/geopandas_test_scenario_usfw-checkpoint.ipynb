### Downloading some practice data:

#### I downloaded data on critical species habitat from the US Fish & Wildlife Service's data web portal, the Environmental Conservation Online System (ECOS). I selected the first zip file [here](https://ecos.fws.gov/ecp/report/table/critical-habitat.html) (last updated July 12, 2022).  

- I put the zip file in this notebook directory. It contains containing two shapefiles, one for lines, one for polygons, which aggregate all critical habitat shapes for all species (updated 07/12/22).

### Setting the coordinate reference system (CRS):

#### Now we need to set the data to the desired coordinate reference system (CRS). We do this because for all kinds of spatial data, files may come with an unfamiliar CRS or have no CRS at all. These are needed for a mask, a raster, or executing similar geospatial processes. This is general guide to [different CRS types](https://www.nceas.ucsb.edu/sites/default/files/2020-04/OverviewCoordinateReferenceSystems.pdf). 

#### We will use a CRD called Mercator (EPSG: 3857), which is used for tiles from Google Maps, Open Street Maps, and Stamen Maps. We'll use it because we want to overlay shapefile polygons onto a Google basemap with the contextily package.

#### Set the CRS to ESPG 3857. This code may take a minute to run:

```python
# Set the CRS for the geometry of all species critical habitats
gdf_3857 = gdf.to_crs(epsg=3857)
print(gdf_3857.crs)
```

### Plotting the shapefiles on a basemap:

#### Now we need to use matplotlib's plot() function to make the polygon color represent each species.

 - We will name the dataframe which we want to plot, gdf_3857 in this case, then plot() and add arguments and supplemental plot structure changes as you go.

```python
gdf_3857.plot(column='comname',
             figsize=(20,20))
```

#### These polygons are just floating in space, and so we'll add in the basemap for some geographical context:

- We will use an argument in the plot function, setting the column = 'comname', which is a column within the gdf_3857 geodataframe that specifies the common name for the species in that row. This argument sets a unique color to each common name, which will help us tell the difference between each speciesâ€™ habitat on the map, even if 1 speciesâ€™ habitat is composed of multiple polygons.

```python
crit_hab_polys = gdf_3857.plot(column='comname',
             figsize=(20,20))
ctx.add_basemap(crit_hab_polys) # add in the basemap
```

### Zooming in on a region of interest:

- Since the basemap within the contextily package is of the entire world, we need to specify the x-limitations and y-limitations for our map so we can zoom in on the United States/somewhere within the US to best understand the critical habitiat data. 

- This took some guessing. The default x and y units were in the millions, so we specified units in millions, too. The coordinate reference systems is designed with positive values for North and East, and negative values for South and West. Since the United States is north of the equator, we should have positive values for the min and max y. 

- For value magnitude, you can look at the map for a starting point and play around with different numbers until getting the desired ROI.

```python
crit_hab_polys = gdf_3857.plot(column='comname',
             figsize=(20,20))
ctx.add_basemap(crit_hab_polys) # add in the basemap

# Set the axis limits to zoom in on just the lower 48 states, rather than viewing the entire world:  
plt.ylim([2350000,6350000])
plt.xlim([-14000000,-6000000]);
```
