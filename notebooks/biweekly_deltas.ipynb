### Evaluating the differences in conductivity between sites up- and downstream of injection wells

```python
# LIBRARIES
#import pandas library for working with tabular data
import os
os.getcwd()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
from scipy.stats import kde
from matplotlib import cm
import datetime
import matplotlib.dates as mdates
import matplotlib.cbook as cbook
#import regular expresson (regex)
import re
#check pandas version
pd.__version__
```

```python
# read from csv 
biweek = pd.read_csv("/home/millieginty/Documents/git-repos/ccbor/analyses/biweekly_differences.csv", parse_dates=['Date'])

biweek.head()
```

```python
# West Branch Alum Creek deltas (Turkey Run)

fig, ax = plt.subplots(figsize=(15, 10))

# Varaibles
small = 15
med = 20
large = 25
linew = 3

# Plotting things 

ax.plot(biweek.Date, biweek.WBAdiff1, c = '#3182bd', label='# Downstream 1 - Upstream', linewidth=linew) 
ax.plot(biweek.Date, biweek.WBAdiff2, c = 'black', label='# Downstream 2 - Upstream', linewidth=linew, linestyle = 'dashed')

ax.tick_params(axis='y', labelsize= med)
ax.set_xticklabels(biweek.Date, rotation = 85, size= med)
ax.set_ylabel('Conductivity (uS)', size= med)
ax.set_title('West Branch Alum Creek conductivity differences', size= large)
plt.axhline(y = 0, color = 'black', linestyle = '-')

plt.legend(bbox_to_anchor=(1,1), loc="upper left", fontsize = med)

ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%b'))
for label in ax.get_xticklabels(which='major'):
    label.set(rotation=30, horizontalalignment='right')

plt.show()
```

```python
# Alum Creek deltas 

fig, ax = plt.subplots(figsize=(15, 10))

# Varaibles
small = 15
med = 20
large = 25
linew = 3

# Plotting things 

ax.plot(biweek.Date, biweek.Aldiff1, c = '#3182bd', label='# Intermediate 1 - Upstream', linewidth=linew) 
ax.plot(biweek.Date, biweek.Aldiff2, c = 'black', label='# Intermediate 2 - Upstream', linewidth=linew, linestyle = 'dashed')
ax.plot(biweek.Date, biweek.Aldiff3, c = 'red', label='# Downstream - Upstream', linewidth=linew, linestyle = 'dotted')

ax.tick_params(axis='y', labelsize= med)
ax.set_xticklabels(biweek.Date, rotation = 85, size= med)
ax.set_ylabel('Conductivity (uS)', size= med)
ax.set_title('West Branch Alum Creek conductivity differences', size= large)
plt.axhline(y = 0, color = 'black', linestyle = '-')

plt.legend(bbox_to_anchor=(1,1), loc="upper left", fontsize = med)

ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%b'))
for label in ax.get_xticklabels(which='major'):
    label.set(rotation=30, horizontalalignment='right')

plt.show()
```
